# Database Quiz - 2

---

### 1. What is the name of the official MongoDB driver for Python?  
- [ ] MongoPython  
- [ ] PyMongoDB  
- [ ] MongoDB-Py  
- [x] PyMongo  

**Explanation**: PyMongo is the official MongoDB driver for Python. It is a Python library that allows you to interact with MongoDB from Python, providing a simple way to connect, query data, and perform CRUD operations.

---

### 2. How do you connect to MongoDB using PyMongo in Python?  
- [ ] By using the `connect()` method  
- [ ] By using the `PyMongo()` method  
- [x] By using the `MongoClient()` method  
- [ ] By using the `MongoDB()` method  

**Explanation**: To connect to MongoDB using PyMongo in Python, you use the `MongoClient()` method. It takes a connection string as an argument to specify the host and port of the MongoDB server.

---

### 3. How do you access a collection in MongoDB using PyMongo in Python?  
- [ ] By using the `collection()` method  
- [ ] By using the `get_collection()` method  
- [ ] By using the `get_database()` method  
- [x] By using the `database[collection]` syntax  

**Explanation**: To access a collection in MongoDB using PyMongo, you use the `database[collection]` syntax. This allows you to perform CRUD operations on the collection.

---

### 4. How do you insert a document into a collection in MongoDB using PyMongo in Python?  
- [x] By using the `insert_one()` method  
- [ ] By using the `insert()` method  
- [ ] By using the `save()` method  
- [ ] By using the `add()` method  

**Explanation**: The `insert_one()` method is used to insert a document into a collection in MongoDB. It takes a document as an argument and returns an `InsertOneResult` object.

---

### 5. How do you query documents from a collection in MongoDB using PyMongo in Python?  
- [x] By using the `find()` method  
- [ ] By using the `query()` method  
- [ ] By using the `get()` method  
- [ ] By using the `select()` method  

**Explanation**: The `find()` method is used to query documents from a collection. It returns a cursor to iterate over the matching documents.

---

### 6. What is MongoDB?  
- [ ] A relational database management system  
- [x] A document-oriented database management system  
- [ ] A graph database management system  
- [ ] A key-value store database management system  

**Explanation**: MongoDB is a document-oriented database management system that stores data as documents in BSON (Binary JSON). It supports rich data types like arrays and nested structures.

---

### 7. What is a collection in MongoDB?  
- [ ] A database  
- [ ] A table  
- [x] A set of documents  
- [ ] A set of rows  

**Explanation**: A collection in MongoDB is a set of documents. It is equivalent to a table in relational databases but offers a more flexible data model.

---

### 8. What is a document in MongoDB?  
- [ ] A row  
- [ ] A tuple  
- [ ] An instance of data  
- [x] A JSON-like structure  

**Explanation**: A document in MongoDB is a JSON-like structure representing an instance of data. It can include fields with various data types and nested documents.

---

### 9. What is the default storage engine for MongoDB?  
- [ ] RocksDB  
- [x] WiredTiger  
- [ ] InnoDB  
- [ ] MyISAM  

**Explanation**: WiredTiger is the default storage engine for MongoDB. It provides high performance, scalability, and data compression, along with features like document-level locking.

---

### 10. What is the purpose of the MongoDB aggregation framework?  
- [x] To perform complex data processing on collections  
- [ ] To enforce data consistency and integrity  
- [ ] To manage transactions and rollbacks  
- [ ] To define indexing strategies  

**Explanation**: The MongoDB aggregation framework allows for complex data processing, such as grouping, filtering, and transforming data. It uses a pipeline-based interface to build data processing pipelines.
