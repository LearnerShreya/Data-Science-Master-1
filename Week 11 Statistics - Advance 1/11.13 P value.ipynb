# üìå **P-Value (Probability Value)**

The **p-value** is a fundamental concept in statistical hypothesis testing. It quantifies the probability of obtaining a test statistic as extreme or more extreme than the observed value, assuming that the **null hypothesis (H‚ÇÄ)** is true. The p-value helps us decide whether to **reject or fail to reject** the null hypothesis.

---

## üìù **How to Interpret the P-Value?**

### ‚úÖ **1. Decision Rule**
- If **p ‚â§ Œ±** (where **Œ±** is the significance level, typically **0.05** or **0.01**):
  - There is **strong evidence against** the null hypothesis (**H‚ÇÄ**).
  - We **reject H‚ÇÄ** (the result is **statistically significant**).

- If **p > Œ±**:
  - There is **weak evidence against** the null hypothesis (**H‚ÇÄ**).
  - We **fail to reject H‚ÇÄ** (not enough evidence to conclude significance).

## üìù **Example: One-Tailed Z-Test**

A company claims the average weight of its product is **500g**.  
A sample of **30** items has an average weight of **490g**, with a standard deviation of **15g**.  
We test:

- **Null hypothesis (H‚ÇÄ):** The mean weight is **500g**.
- **Alternative hypothesis (H‚ÇÅ):** The mean weight is **less than 500g** (one-tailed test).
- **Significance level:** Œ± = 0.05

### üîπ **Step 1: Compute the Z-Score**
**Z = (490 - 500) / (15 / ‚àö30) = -10 / 2.74 = -3.65**

### üîπ **Step 2: Find the P-Value**

**P(Z < -3.65) ‚âà 0.00013**

Since **0.00013 < 0.05**, we **reject H‚ÇÄ**.  
üîπ **Conclusion:** The mean weight is significantly **less than 500g**.
