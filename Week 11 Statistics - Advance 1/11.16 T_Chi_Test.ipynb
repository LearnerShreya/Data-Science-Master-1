# üìå **T-Test**

A **t-test** is a statistical hypothesis test used to determine whether there is a significant difference between the means of two groups or between a sample mean and a population mean. It is commonly used when the sample size is small, and the population standard deviation is unknown.

---

## üìù **How to Interpret the T-Test?**

### ‚úÖ **1. Decision Rule**
- If **p ‚â§ Œ±** (where **Œ±** is the significance level, typically **0.05** or **0.01**):
  - There is **strong evidence against** the null hypothesis (**H‚ÇÄ**).
  - We **reject H‚ÇÄ** (the result is **statistically significant**).

- If **p > Œ±**:
  - There is **weak evidence against** the null hypothesis (**H‚ÇÄ**).
  - We **fail to reject H‚ÇÄ** (not enough evidence to conclude significance).

---

## üî¢ **Formula for T-Test Calculation**

The t-test is based on the **t-statistic**, which is calculated using the following formula:

$$
t = \frac{\bar{X} - \mu}{\frac{s}{\sqrt{n}}}
$$

Where:
- **\( \bar{X} \)** = Sample mean  
- **\( \mu \)** = Population mean under **H‚ÇÄ**  
- **\( s \)** = Sample standard deviation  
- **\( n \)** = Sample size  

After finding the **t-statistic**, we determine the **p-value** from the t-distribution table.

---

## üìù **Example: One-Sample T-Test**

A researcher claims that the average weight of a species of fish is **300g**. A sample of **25 fish** has an average weight of **290g**, with a standard deviation of **10g**. We test:

- **Null hypothesis (H‚ÇÄ):** The mean weight is **300g**.
- **Alternative hypothesis (H‚ÇÅ):** The mean weight is not **300g** (two-tailed test).
- **Significance level:** Œ± = 0.05

### üîπ **Step 1: Compute the T-Statistic**
$$
t = \frac{290 - 300}{\frac{10}{\sqrt{25}}} = \frac{-10}{2} = -5
$$

### üîπ **Step 2: Find the P-Value**
Using the **t-table** for **df = 24** (degrees of freedom),  
$$
 p \approx 0.0001
$$

Since **0.0001 < 0.05**, we **reject H‚ÇÄ**.  
üîπ **Conclusion:** The average weight is significantly different from **300g**.

# üìå **Chi-Square Test**

A **chi-square test** is a statistical test used to determine if there is a significant association between categorical variables. It compares the observed frequency in each category to the expected frequency under the assumption that no association exists.

---

## üìù **How to Interpret the Chi-Square Test?**

### ‚úÖ **1. Decision Rule**
- If **p ‚â§ Œ±** (where **Œ±** is the significance level, typically **0.05** or **0.01**):
  - There is **strong evidence against** the null hypothesis (**H‚ÇÄ**).
  - We **reject H‚ÇÄ** (the result is **statistically significant**).

- If **p > Œ±**:
  - There is **weak evidence against** the null hypothesis (**H‚ÇÄ**).
  - We **fail to reject H‚ÇÄ** (not enough evidence to conclude significance).

---

## üî¢ **Formula for Chi-Square Test Calculation**

The chi-square statistic is calculated using the formula:

$$
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
$$

Where:
- **\( O_i \)** = Observed frequency for category \( i \)
- **\( E_i \)** = Expected frequency for category \( i \)

The chi-square statistic follows a **chi-square distribution** with degrees of freedom (**df**) calculated as:

$$
df = (r - 1)(c - 1)
$$

Where:
- **r** = Number of rows
- **c** = Number of columns

We then find the **p-value** from the chi-square distribution table.

---

## üìù **Example: Chi-Square Test for Independence**

A study wants to know if there is an association between gender and preference for a product. The data collected is as follows:

| Gender     | Product A | Product B | Total |
|-------------|-----------|-----------|-------|
| Male        | 30        | 10        | 40    |
| Female      | 20        | 20        | 40    |
| **Total**   | 50        | 30        | 80    |

We test:

- **Null hypothesis (H‚ÇÄ):** There is no association between gender and product preference.
- **Alternative hypothesis (H‚ÇÅ):** There is an association between gender and product preference.
- **Significance level:** Œ± = 0.05

### üîπ **Step 1: Calculate the Expected Frequencies**
For example, for the Male-Product A cell:
$$
E_{\text{Male, A}} = \frac{50 \times 40}{80} = 25
$$
Repeat for all cells to get expected frequencies.

### üîπ **Step 2: Compute the Chi-Square Statistic**
$$
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
$$

### üîπ **Step 3: Find the P-Value**
Using the **chi-square table**, for **df = 1** (since \( (2 - 1)(2 - 1) = 1 \)), we find the **p-value**.
