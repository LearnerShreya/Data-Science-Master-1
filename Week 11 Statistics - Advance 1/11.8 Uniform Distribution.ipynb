# **Uniform Distribution**

A **Uniform Distribution** is a probability distribution where all outcomes are equally likely. There are two types:

1. **Discrete Uniform Distribution** (for finite outcomes) `PMF`
2. **Continuous Uniform Distribution** (for infinite outcomes in a range) `PDF`

### **Key Points**:
- Every value in a **uniform distribution** has an equal chance of happening.
- In the discrete type, this applies to a finite set of numbers.
- In the continuous type, this applies to all real numbers between two limits.

---

## **Discrete Uniform Distribution**

### **Definition**:
In a **discrete uniform distribution**, there are a set number of possible outcomes, and each one has the same probability.

### **Formula**:
If there are **n** outcomes (e.g., 1, 2, 3, ..., n), the probability of each outcome is:

$$
P(X = x) = \frac{1}{n}
$$

Where:
- **P(X = x)** = Probability of outcome **x**
- **n** = Number of possible outcomes

### **Example**:
If you roll a fair die, there are 6 possible outcomes (1, 2, 3, 4, 5, 6), and the probability of rolling any one of these is:

$$
P(X = x) = \frac{1}{6}
$$

---

## **Continuous Uniform Distribution**

### **Definition**:
In a **continuous uniform distribution**, the probability is evenly spread across an interval [a, b] (e.g., between 0 and 10). Every point in this interval has the same probability density.

### **Probability Density Function (PDF)**:
For a continuous uniform distribution over the interval [a, b], the formula for the probability density is:

$$
f(x) = \frac{1}{b - a}, \quad \text{where } a \leq x \leq b
$$

Where:
- **f(x)** = Probability density at value **x**
- **a** = Lower bound of the interval
- **b** = Upper bound of the interval

### **Example**:
If you randomly pick a number between 0 and 10, the probability density function would be:

$$
f(x) = \frac{1}{10}, \quad 0 \leq x \leq 10
$$

This means the probability of picking any number between 0 and 10 is the same.

---

## **Cumulative Distribution Function (CDF)**

The **Cumulative Distribution Function (CDF)** tells us the probability that a random variable is less than or equal to a certain value.

### **For Continuous Uniform Distribution**:
The CDF for a continuous uniform distribution over the interval [a, b] is:

$$
F(x) = \frac{x - a}{b - a}, \quad \text{where } a \leq x \leq b
$$

Where:
- **F(x)** = Cumulative probability up to point **x**
- **a** = Lower bound
- **b** = Upper bound
- **x** = Value for which we calculate the CDF

### **Example**:
If you want the cumulative probability for a number between 0 and 10, the CDF is:

$$
F(x) = \frac{x}{10}, \quad 0 \leq x \leq 10
$$

## **Mean, Variance, and Standard Deviation**

### **1. Mean (Expected Value)**:
The **mean** (or expected value) is the "average" of all possible outcomes.

For both **discrete** and **continuous** uniform distributions, the mean is:

$$
\mu = \frac{a + b}{2}
$$

Where:
- **a** = Lower bound of the range
- **b** = Upper bound of the range

### **2. Variance**:
Variance tells us how spread out the values are from the mean. For a **continuous uniform distribution**:

$$
Var(X) = \frac{(b - a)^2}{12}
$$

Where:
- **Var(X)** = Variance

### **3. Standard Deviation**:
Standard deviation is the square root of variance, showing the spread of values.

For a **continuous uniform distribution**:

$$
\sigma = \frac{b - a}{\sqrt{12}}
$$

## **Summary of Key Formulas**

1. **Discrete Uniform Distribution (PMF)**:
$$
P(X = x) = \frac{1}{n}, \quad x \in \{1, 2, 3, \dots, n\}
$$

2. **Continuous Uniform Distribution (PDF)**:
$$
f(x) = \frac{1}{b - a}, \quad a \leq x \leq b
$$

3. **Cumulative Distribution Function (CDF)**:
$$
F(x) = \frac{x - a}{b - a}, \quad a \leq x \leq b
$$

4. **Mean (Expected Value)**:
$$
\mu = \frac{a + b}{2}
$$

5. **Variance**:
$$
Var(X) = \frac{(b - a)^2}{12}
$$

6. **Standard Deviation**:
$$
\sigma = \frac{b - a}{\sqrt{12}}
$$

### **Real-World Applications of Uniform Distribution**:
- **Random Sampling**: When selecting samples where each item has an equal chance of being selected.
- **Cryptography**: Ensures that each key is equally likely.
- **Simulation Models**: Used in Monte Carlo simulations to simulate random outcomes.

---

## **Conclusion**:
Uniform distribution is useful for modeling situations where all outcomes are equally likely. The key formulas you need to know for **discrete** and **continuous** uniform distributions are simple and easy to apply in real-world problems.
