# **Normal (Gaussian) Distribution**

## **Introduction**
The **Normal Distribution**, also known as the **Gaussian Distribution**, is a continuous probability distribution that is widely used in statistics and machine learning. It describes data that clusters around a central value with symmetric spread on both sides.

This distribution appears in numerous **natural and social phenomena**, such as heights, IQ scores, blood pressure, stock returns, and errors in measurement.

---

## **Probability Density Function (PDF)**  
The probability density function (PDF) of a normal distribution is given by:

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

where:  
- **Î¼ (mu)** = Mean (center of the distribution)  
- **Ïƒ (sigma)** = Standard deviation (spread of the distribution)  
- **ÏƒÂ²** = Variance  
- **e** = Eulerâ€™s number (~2.718)  

The **normal distribution curve** is **bell-shaped** and **symmetric** around the mean **Î¼**.

### **Interpretation of PDF**  
The **PDF represents the likelihood** of a random variable **X** taking a specific value. Since this is a continuous distribution, the probability of **exactly one value** is **zero**. Instead, we calculate probabilities over a **range of values** using the **area under the curve**.

## **Cumulative Distribution Function (CDF)**
The **cumulative distribution function (CDF)** represents the probability that a random variable **X** takes a value **less than or equal to** a given number **x**:

$$
F(x) = \frac{1}{2} \left[ 1 + \text{erf} \left( \frac{x - \mu}{\sigma \sqrt{2}} \right) \right]
$$

where **erf** is the **error function**.

### **Interpretation of CDF**  
- The **CDF is a monotonically increasing function** between **0 and 1**.  
- **F(x) = P(X â‰¤ x)** gives the probability that **X** is **less than or equal to** x.  
- The CDF is used to find probabilities for **ranges of values**.

## **Standard Normal Distribution**
A **Standard Normal Distribution** is a special case where:
- **Mean Î¼ = 0**  
- **Standard Deviation Ïƒ = 1**  

To convert any normal variable **X** to the standard normal form, use the **Z-score transformation**:

$$
Z = \frac{X - \mu}{\sigma}
$$

where **Z** represents the number of standard deviations a value **X** is from the mean.

### **Z-Score Table & Applications**
- A **Z-score table** is used to find probabilities for standard normal values.
- **Example:** If Z = **-1.5**, the probability of getting a value **less than** that point can be found using the **Z-table**.
- **Applications:** Hypothesis testing, confidence intervals, anomaly detection.

## **Normal Approximation to Binomial**
For a binomial distribution with **large n**, the **normal distribution** can be used as an approximation when:

$$
n \cdot p > 5 \quad \text{and} \quad n \cdot (1 - p) > 5
$$

The normal approximation to binomial uses:

$$
X \sim N(np, \sqrt{np(1-p)})
$$

where **np** is the mean and **âˆš(np(1-p))** is the standard deviation.

## **Visual Representation**
The **normal distribution curve** can be plotted in Python using Matplotlib:


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define mean and standard deviation
mu, sigma = 0, 1  

# Generate x values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Compute PDF
y = norm.pdf(x, mu, sigma)

# Plot the curve
plt.figure(figsize=(8,5))
plt.plot(x, y, color='blue', linewidth=2)
plt.title("Standard Normal Distribution")
plt.xlabel("X values")
plt.ylabel("Probability Density")
plt.grid()
plt.show()
```
