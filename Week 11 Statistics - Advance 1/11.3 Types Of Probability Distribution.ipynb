# 🌟 Types of Probability Distributions: PMF & PDF

Probability distributions are mathematical functions that describe the likelihood of different outcomes in an experiment. Each type of distribution has its own **Probability Mass Function (PMF)** for discrete variables or **Probability Density Function (PDF)** for continuous variables. Below are the most common types of distributions, along with their formulas and key characteristics:

---

## 1. Normal (Gaussian) Distribution 🌐  
### 📌 **Definition:**  
The **Normal Distribution**, also known as the **Gaussian Distribution**, is the most widely used continuous probability distribution. It is characterized by a **bell-shaped curve** that is symmetrical around the mean. It describes many natural phenomena and measurement errors.

### 📌 **PDF Formula:**  
For a continuous random variable **X**, the **PDF** is:  

**f(x) = (1 / √(2πσ²)) e<sup>-(x - μ)² / 2σ²</sup>**

- **μ** = Mean of the distribution  
- **σ** = Standard deviation  
- **π** = Pi (≈ 3.14159)  
- **e** = Euler's number (≈ 2.71828)

### 📌 **Properties:**  
- Symmetric around the mean **μ**  
- **68%** of the data lies within one standard deviation  
- **95%** lies within two standard deviations  
- **99.7%** lies within three standard deviations  

### 📌 **Example:**  
The distribution of heights of people is often **normally distributed**. 📏

---

## 2. Bernoulli Distribution 🎯  
### 📌 **Definition:**  
The **Bernoulli Distribution** models binary outcomes, such as **success/failure** or **yes/no**. It is the simplest discrete distribution, with only two possible outcomes.

### 📌 **PMF Formula:**  
For a random variable **X** representing the outcome of a Bernoulli trial:

**P(X = x) = p<sup>x</sup> (1 - p)<sup>1 - x</sup>**, where **x ∈ {0, 1}**

- **p** = Probability of success (0 ≤ **p** ≤ 1)  
- **1 - p** = Probability of failure  

### 📌 **Properties:**  
- Only two possible outcomes: **0** (failure) or **1** (success)  
- **E(X) = p** (Expected value)  
- **Var(X) = p(1 - p)** (Variance)

### 📌 **Example:**  
A **coin toss** with **heads** as success and **tails** as failure. 🪙

---

## 3. Uniform Distribution 🔳  
### 📌 **Definition:**  
The **Uniform Distribution** describes a scenario where all outcomes in a specified range are equally likely. It can be either **discrete** or **continuous**.

### 📌 **PDF Formula:**  
For a continuous random variable **X** uniformly distributed over the interval **[a, b]**:

**f(x) = 1 / (b - a)**, for **a ≤ x ≤ b**

For a discrete uniform distribution, the PMF is:

**P(X = x) = 1 / (b - a + 1)**, for **x ∈ {a, a+1, a+2, ..., b}**

### 📌 **Properties:**  
- All outcomes in the range are equally probable  
- **E(X) = (a + b) / 2** (Expected value)  
- **Var(X) = (b - a)² / 12** (Variance)

### 📌 **Example:**  
Rolling a fair **six-sided die** is a **discrete uniform distribution**. 🎲

---

## 4. Log-Normal Distribution 📈  
### 📌 **Definition:**  
The **Log-Normal Distribution** is a continuous distribution where the **logarithm** of the variable is normally distributed. It is typically used to model **positively skewed** data.

### 📌 **PDF Formula:**  
For a continuous random variable **X**, the **PDF** is:

**f(x) = (1 / (xσ√(2π))) e<sup>-(ln(x) - μ)² / 2σ²</sup>**, for **x > 0**

- **μ** = Mean of the natural logarithm of **X**  
- **σ** = Standard deviation of the natural logarithm of **X**

### 📌 **Properties:**  
- Skewed to the right (positive skew)  
- **Used to model** data that can’t be negative, such as **stock prices** and **income**

### 📌 **Example:**  
Stock market returns often follow a **log-normal distribution**. 📉

---

## 5. Poisson Distribution 🧑‍🔬  
### 📌 **Definition:**  
The **Poisson Distribution** models the number of **events** occurring within a fixed interval of time or space. The events occur **independently** and at a constant rate.

### 📌 **PMF Formula:**  
For a random variable **X** representing the number of events in a Poisson distribution with mean **λ**:

**P(X = x) = (λ<sup>x</sup> e<sup>-λ</sup>) / x!**, where **x = 0, 1, 2, ...**

- **λ** = Average number of events in the given time/space

### 📌 **Properties:**  
- **Mean = Variance = λ**  
- Models rare events (e.g., accidents, calls)  

### 📌 **Example:**  
The number of **accidents** at a traffic intersection in a given time period. 🚗

---

## 6. Power Law Distribution ⚡  
### 📌 **Definition:**  
The **Power Law Distribution** is a heavy-tailed distribution used to model phenomena where the frequency of events decreases polynomially with the size of the event.

### 📌 **PDF Formula:**  
For a random variable **X**, the **Power Law PDF** is:

**f(x) = Cx<sup>-α</sup>**, for **x ≥ x₀**

- **α** = Exponent (α > 1)  
- **C** = Normalizing constant  
- **x₀** = Minimum value of **X**

### 📌 **Properties:**  
- Heavy-tailed, meaning rare, high-impact events are still likely to occur  
- Often used to model **internet traffic**, **wealth distribution**, and **natural disasters**

### 📌 **Example:**  
Income distribution follows a **Power Law Distribution**, with a small number of individuals holding most wealth. 💸

---

## 7. Binomial Distribution 🎯  
### 📌 **Definition:**  
The **Binomial Distribution** models the number of **successes** in **n** independent Bernoulli trials, each with the same probability of success.

### 📌 **PMF Formula:**  
For a random variable **X** representing the number of successes in **n** trials, with probability **p** of success:

**P(X = x) = (nCx) p<sup>x</sup> (1 - p)<sup>n - x</sup>**, for **x = 0, 1, 2, ..., n**

- **nCk** = Binomial coefficient  

### 📌 **Properties:**  
- **Mean = np**, **Variance = np(1 - p)**  
- Models **yes/no** experiments, like **coin flips** or **successes in trials**

### 📌 **Example:**  
The number of **heads** in 10 coin flips follows a **Binomial Distribution**. 🪙

---

## Key Takeaways 🌟  
- **Normal Distribution**: Continuous, bell-shaped, symmetric  
- **Bernoulli Distribution**: Discrete, two possible outcomes  
- **Uniform Distribution**: Equal probability across the range  
- **Log-Normal Distribution**: Skewed, often used for financial data  
- **Poisson Distribution**: Models countable events over intervals  
- **Power Law Distribution**: Heavy-tailed, for rare but impactful events  
- **Binomial Distribution**: Discrete, for fixed trials and binary outcomes

---

## 💡 Real-World Applications  
- **Normal Distribution**: Heights, test scores, measurement errors  
- **Bernoulli**: Coin toss, pass/fail outcomes  
- **Uniform**: Rolling dice, random sampling  
- **Log-Normal**: Stock prices, income distributions  
- **Poisson**: Accidents, phone calls, emails  
- **Power Law**: Wealth inequality, internet traffic  
- **Binomial**: Quality control, clinical trials
