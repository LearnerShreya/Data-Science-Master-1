# 📊 Binomial Distribution

## 🔹 What is it?
The **Binomial Distribution** models the number of **successes** in a fixed number of **independent** trials, where each trial has only **two possible outcomes**:
- ✅ **Success** (1)
- ❌ **Failure** (0)

Common examples include:
- Flipping a coin multiple times 🪙  
- Checking for defective products in a batch 🔍  
- Predicting the success rate of a marketing campaign 📢  

---

## 🔹 Formula  
The probability of getting exactly **k** successes in **n** trials is:

**P(X = k) = C(n, k) p<sup>k</sup> (1 - p)<sup>(n - k)</sup>**

where:
- **n** = Total number of trials  
- **k** = Number of successes  
- **p** = Probability of success in one trial  
- **1 - p** = Probability of failure in one trial  
- **C(n, k)** = Combination formula:

  **C(n, k) = n! / (k!(n - k)!)**

---

## 🔹 Key Properties
| Property | Formula |
|----------|---------|
| **Mean (Expected Value)** | **μ = np** |
| **Variance** | **σ² = np(1 - p)** |
| **Standard Deviation** | **σ = √(np(1 - p))** |

---

## 🔹 Example: Coin Toss 🎲
Suppose you flip a **fair coin** **5 times** (**n = 5**) and want to find the probability of getting **exactly 3 heads** (**k = 3**).  

- Since the coin is fair, the probability of heads is **p = 0.5**.  
- Using the binomial formula:

  **P(X = 3) = C(5, 3) (0.5)<sup>3</sup> (0.5)<sup>2</sup>**  

  **= (5! / (3!(5 - 3)!)) × (0.5)<sup>5</sup>**  

  **= (5 × 4) / (2 × 1) × 0.03125**  

  **= 10 × 0.03125**  

  **= 0.3125 (or 31.25%)**  

So, the probability of getting exactly **3 heads in 5 flips** is **31.25%**.

## 🔹 Real-World Applications  
✅ **Quality Control** 🏭 (Checking defective items in production)  
✅ **Medical Studies** 🏥 (Success rate of a treatment)  
✅ **Sports Analytics** ⚽ (Probability of a player scoring goals)  
✅ **Marketing** 📈 (Conversion rate of advertisements)  

---

## 💡 Key Takeaways  
✔ Use the Binomial Distribution when you have **fixed trials, independent events, and only two outcomes per trial**.  
✔ If **p = 0.5**, the distribution is **symmetric** (like a fair coin).  
✔ If **p** is small or large, the distribution becomes **skewed**.
