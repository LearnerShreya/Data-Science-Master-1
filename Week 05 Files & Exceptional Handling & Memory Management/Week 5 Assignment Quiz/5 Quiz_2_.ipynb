# Exception Handling Quiz

### 1. What is the purpose of the `try` and `except` keywords in Python?
- **Options:**
  - A. To handle errors and exceptions in a program.
  - B. To perform conditional operations.
  - C. To create user-defined functions.
  - D. To import modules.

- **Correct Answer:** A. To handle errors and exceptions in a program.

- **Explanation:** The `try` and `except` keywords are used to handle errors and exceptions in a program. The `try` block contains the code that might raise an exception, and the `except` block contains the code that will handle the exception if it occurs.

---

### 2. What happens if an exception is raised in the `try` block?
- **Options:**
  - A. The `except` block will be executed.
  - B. The program will continue to run.
  - C. The program will stop executing.
  - D. None of the above.

- **Correct Answer:** A. The `except` block will be executed.

- **Explanation:** If an exception is raised in the `try` block, the program will immediately jump to the `except` block to handle the exception. If no exception is raised, the code in the `except` block will not be executed.

---

### 3. What happens if no exception is raised in the `try` block?
- **Options:**
  - A. The `except` block will be executed.
  - B. The program will continue to run.
  - C. The program will stop executing.
  - D. None of the above.

- **Correct Answer:** B. The program will continue to run.

- **Explanation:** If no exception is raised in the `try` block, the code in the `except` block will not be executed, and the program will continue to run normally.

---

### 4. What is the use of the `finally` keyword in exception handling?
- **Options:**
  - A. To handle exceptions and errors.
  - B. To specify a block of code that will be executed regardless of whether an exception was raised or not.
  - C. To create user-defined functions.
  - D. None of the above.

- **Correct Answer:** B. To specify a block of code that will be executed regardless of whether an exception was raised or not.

- **Explanation:** The `finally` keyword can be used to specify a block of code that will always be executed, even if an exception is raised in the `try` block or if an error occurs in the `except` block.

---

### 5. What is a custom exception in Python?
- **Options:**
  - A. A built-in exception.
  - B. A user-defined exception.
  - C. Both A and B.
  - D. None of the above.

- **Correct Answer:** B. A user-defined exception.

- **Explanation:** A custom exception in Python is a user-defined exception that can be raised and caught in the same way as built-in exceptions.

---

### 6. How can you create a custom exception in Python?
- **Options:**
  - A. By using the `raise` statement.
  - B. By using the `assert` statement.
  - C. Both A and B.
  - D. None of the above.

- **Correct Answer:** A. By using the `raise` statement.

- **Explanation:** To create a custom exception in Python, you need to define a new exception class that inherits from the built-in `Exception` class. You can then raise the custom exception using the `raise` statement.
