# Exception Handling Quiz - 2

### 1. _____ is a type of error produced in Python programs.
- **Options:**
  - A. Compile-time error
  - B. Run-time error
  - C. Both (A) and (B)
  - D. None of the above

- **Correct Answer:** C. Both (A) and (B)

- **Explanation:** A compile-time error generally refers to errors in semantics or syntax, which can be fixed during development. A runtime error occurs during code execution and cannot be detected by the compiler.

---

### 2. What exception is raised when a name is not found in the local or global namespace?
- **Options:**
  - A. TypeError
  - B. NameError
  - C. ValueError
  - D. ImportError

- **Correct Answer:** B. NameError

- **Explanation:** `NameError` is raised when a name is not found in the local or global namespace.

---

### 3. What exception is raised when an attempt is made to divide a number by zero?
- **Options:**
  - A. ZeroDivisionError
  - B. KeyError
  - C. IndexError
  - D. TypeError

- **Correct Answer:** A. ZeroDivisionError

- **Explanation:** `ZeroDivisionError` is raised when an attempt is made to divide a number by zero.

---

### 4. What exception is raised when a file or directory is requested but does not exist?
- **Options:**
  - A. SyntaxError
  - B. FileNotFoundError
  - C. AttributeError
  - D. ImportError

- **Correct Answer:** B. FileNotFoundError

- **Explanation:** `FileNotFoundError` is raised when a file or directory is requested but does not exist.

---

### 5. What exception is raised when an attribute reference or assignment fails?
- **Options:**
  - A. ZeroDivisionError
  - B. KeyError
  - C. AttributeError
  - D. TypeError

- **Correct Answer:** C. AttributeError

- **Explanation:** `AttributeError` is raised when an attribute reference or assignment fails.
