# Quiz : Files Quiz

## Python and File Handling Questions and Answers

### 1. To open a file `c:\scores.txt` for reading, we use:
- **Options:**
  - A. `.infile = open(“c:\scores.txt”, “r”)`
  - B. `.infile = open(“c:\\scores.txt”, “r”)`
  - C. `infile = open(file = “c:\scores.txt”, “r”)`
  - D. `infile = open(file = “c:\\scores.txt”, “r”)`

- **Answer:** B. `.infile = open(“c:\\scores.txt”, “r”)`

- **Explanation:** Double backslashes (`\\`) are required to escape the backslash in file paths. Execute `help(open)` for more details.

### 2. Which of the following statements are true?
- **Options:**
  - A. When you open a file for reading, if the file does not exist, an error occurs.
  - B. When you open a file for writing, if the file does not exist, a new file is created.
  - C. When you open a file for writing, if the file exists, the existing file is overwritten with the new file.
  - D. All of the mentioned.

- **Answer:** D. All of the mentioned.

- **Explanation:** The program will throw an error if the file does not exist in read mode, but in write mode, it creates or overwrites the file.

### 3. Which one of the following is not an attribute of a file?
- **Options:**
  - A. `closed`
  - B. `softspace`
  - C. `rename`
  - D. `mode`

- **Answer:** C. `rename`

- **Explanation:** `rename` is not a file attribute. Other attributes include:
  - `file.closed`: Returns `True` if the file is closed.
  - `file.mode`: Returns the mode in which the file was opened.
  - `file.softspace`: Controls spacing in print output.

### 4. Which of these methods are used to read from a file?
- **Options:**
  - A. `open()`
  - B. `read()`
  - C. `scan()`
  - D. `readFileInput()`

- **Answer:** B. `read()`

- **Explanation:** The `read()` method reads content from a file. It returns `-1` when the end of the file is encountered.

### 5. The logging …….. in Python.
- **Options:**
  - A. `function`
  - B. `module`
  - C. `variable`
  - D. `datatype`

- **Answer:** B. `module`

- **Explanation:** The `logging` module provides ready-to-use functionality for creating log messages and is widely used in Python applications. It is a powerful module that is designed to meet the needs of beginners as well as enterprise teams. It is used by most of the third-party Python libraries, so you can integrate your log messages with the ones from those libraries to produce a homogeneous log for your application.

### 6. Which one of the following is the lowest level of logging?
- **Options:**
  - A. `debug`
  - B. `warning`
  - C. `error`
  - D. `none of the above`

- **Answer:** A. `debug`

### 7. Which of the following is not a logging function in Python?

### 8. Which of these is not involved in the process of debugging?
- **Options:**
  - A. `Fixing`
  - B. `Isolating`
  - C. `Identifying`
  - D. `Testing`

### 9. What is the purpose of the `import` statement in Python?
- **Options:**
  - A. `Import the statement into a Python program`
  - B. `Import the function into a Python program`
  - C. `Import the module into a Python program`
  - D. `None of the above`

- **Answer:** C. `Import the module into a Python program`

- **Explanation:** The `import` statement allows access to all definitions within a module.

- **Explanation:** The import statement is used to import a module into a Python program, allowing you to use its definitions and statements. By using the import statement, you can access all the variables, functions, and classes defined in the imported module. You can then use them in your program as if they were defined locally.

### 10. How can you reload a module in Python?
- **Options:**
  - A. `importlib.reload function`
  - B. `import reload`
  - C. `import logging`
  - D. `None of the above`

- **Answer:** A. `importlib.reload function`

- **Explanation:** `importlib.reload` allows reloading an already imported module, useful during module development.

- **Explanation:** The importlib.reload function reloads a previously imported module, allowing you to pick up changes that have been made to the module's code. This is useful in situations where you are developing a module and need to test changes without restarting your program.

### 11. Which of these classes contains the methods used to write to a file?
- **Options:**
  - A. `FileStream`
  - B. `FileInputStream`
  - C. `BufferedOutputStream`
  - D. `FileBufferStream`

- **Answer:** C. `BufferedOutputStream`

- **Explanation:** In Java, the FileOutputStream and FileWriter classes can be used to write to a file. The BufferedOutputStream class provides a buffering layer on top of an OutputStream, which can improve performance when writing to a file.

- **Explanation:** In Java, `BufferedOutputStream` improves performance when writing to a file by buffering data.
