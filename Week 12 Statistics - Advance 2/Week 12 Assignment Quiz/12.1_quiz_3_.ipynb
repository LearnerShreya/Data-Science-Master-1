1. **Which of the following is NOT an assumption to use ANOVA?**  

   - ✅ **Equidistant data points**  
   - Homogeneity of variance  
   - Independence of observations  
   - Normality of residuals  

   **Explanation:** The assumption of equidistant data points is not required for ANOVA. The primary assumptions are homogeneity of variance, independence of observations, and normality of residuals.

2. **Which type of ANOVA is appropriate when there is only one factor being tested?**  

   - ✅ **One-Way ANOVA**  
   - Two-Way ANOVA  
   - Three-Way ANOVA  
   - Factorial ANOVA  

   **Explanation:** One-Way ANOVA is used when there is only one independent variable (factor) being tested for its effect on a dependent variable.

3. **Which term in the ANOVA table represents the amount of variation due to differences between groups?**  

   - Within Groups  
   - Total  
   - ✅ **Between Groups**  
   - Error  

   **Explanation:** The "Between Groups" term in the ANOVA table represents the amount of variation caused by differences between groups. This is used to determine if there is a significant difference between the group means.

5. **What percentage of the total variation in the response variable is explained by the differences between groups in ANOVA?**  

   - Between Groups Variance / Total Variance  
   - Within Groups Variance / Total Variance  
   - ✅ **Between Groups Variance / (Between Groups Variance + Within Groups Variance)**  
   - Within Groups Variance / (Between Groups Variance + Within Groups Variance)  

   **Explanation:** The proportion of total variation explained by differences between groups is calculated as:  
   \[
   R^2 = \frac{\text{Between Groups Variance}}{\text{Between Groups Variance} + \text{Within Groups Variance}}
   \]  
   This is also known as the **coefficient of determination** in ANOVA.

6. **In a One-Way ANOVA with 4 groups, what degrees of freedom are associated with the Within Groups sum of squares?**  

   - 1  
   - 2  
   - 3  
   - ✅ **4**  

   **Explanation:** In a One-Way ANOVA with \( k \) groups, the degrees of freedom for the **Within Groups sum of squares** is calculated as:  
   \[
   df_{\text{within}} = N - k
   \]  
   where \( N \) is the total number of observations and \( k \) is the number of groups.
