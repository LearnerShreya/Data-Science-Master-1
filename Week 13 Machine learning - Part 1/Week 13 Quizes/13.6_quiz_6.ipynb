### **1. Which of the following is true about nominal encoding?**  

⬜ It is used for ordinal data.  
✅ It assigns numerical values to categorical data.  
⬜ It converts categorical data into binary values.  
⬜ It cannot handle datasets with multiple categorical variables.  

#### **Explanation:**  
✅ **Correct Answer:** Nominal encoding assigns numerical values to categorical data without implying any order.  

⬜ **Used for ordinal data?** Incorrect—ordinal encoding is used for ordered categories.  
⬜ **Converts into binary values?** No, that's binary encoding.  
⬜ **Cannot handle multiple categorical variables?** Incorrect—nominal encoding works for multiple categorical features.
