1. **What is the purpose of Min-Max Scaling?**  

   ⬜ To standardize data using mean and standard deviation  
   ✅ **To normalize data into a range of [0,1]**  
   ⬜ To remove outliers from the data  
   ⬜ To perform feature extraction  

   **Explanation:**  
   - **Correct Answer:** Min-Max Scaling (Normalization) transforms data into a range of [0,1] by subtracting the minimum value and dividing by the range. This is useful for ensuring all features have equal importance in analysis.  
   - **Incorrect Options:**  
     - Standardization (Z-score normalization) uses mean and standard deviation, not Min-Max Scaling.  
     - Min-Max Scaling does not remove outliers; it just rescales the data.  
     - Feature extraction involves transforming data into new features, not normalizing it.

2. **What is Feature Extraction?**  

   ⬜ The process of selecting the most important features from a dataset  
   ✅ **The process of transforming data into a different representation**  
   ⬜ The process of combining multiple datasets into a single dataset  
   ⬜ The process of removing outliers from a dataset  

   **Explanation:**  
   - **Correct Answer:** Feature Extraction involves transforming data into a different representation that is more suitable for analysis, such as combining features or creating new ones.  
   - **Incorrect Options:**  
     - Feature selection picks important features but does not transform data.  
     - Combining datasets is data merging, not feature extraction.  
     - Feature extraction does not focus on outlier removal.

3. **What is PCA (Principal Component Analysis)?**  

   ✅ **A feature extraction technique that transforms data into a lower-dimensional space**  
   ⬜ A classification algorithm used for image recognition  
   ⬜ A clustering algorithm used for unsupervised learning  
   ⬜ An outlier detection technique used for anomaly detection  

   **Explanation:**  
   - **Correct Answer:** PCA reduces dimensionality by finding principal components—linear combinations of original features that explain the most variance.  
   - **Incorrect Options:**  
     - PCA is not a classification algorithm.  
     - PCA does not cluster data but reduces dimensions.  
     - PCA is not designed for anomaly detection.
