

### **Feature Engineering Quiz - 5**  

#### **1. Which of the following encoding techniques is used for categorical variables with no inherent ordering?**  
⬜ Ordinal Encoding  
✅ Label Encoding  
⬜ Target Guided Ordinal Encoding  
⬜ None of the above  

**Explanation:**  
✅ **Correct Answer: Label Encoding** is used for categorical variables **without inherent ordering,** assigning a unique integer to each category.  

⬜ **Ordinal Encoding?** Incorrect—used when categories have a natural order.  
⬜ **Target Guided Ordinal Encoding?** Incorrect—assigns values based on the target variable's mean.  
⬜ **None of the above?** Incorrect—Label Encoding is the correct method.  

---

#### **2. Which encoding technique is used when there is a natural ordering between the categories of a categorical variable?**  
✅ Ordinal Encoding  
⬜ Label Encoding  
⬜ Target Guided Ordinal Encoding  
⬜ None of the above  

**Explanation:**  
✅ **Correct Answer: Ordinal Encoding** assigns numerical values based on the natural order of categories.  

⬜ **Label Encoding?** Incorrect—used for unordered categories.  
⬜ **Target Guided Ordinal Encoding?** Incorrect—based on target variable statistics, not inherent order.  
⬜ **None of the above?** Incorrect—Ordinal Encoding is the right method.  

---

#### **3. Covariance is a measure of:**  
⬜ The strength of the linear relationship between two variables  
⬜ The degree of association between two variables  
⬜ The direction of the relationship between two variables  
✅ All of the above  

**Explanation:**  
✅ **Correct Answer: All of the above**—Covariance measures the strength, direction, and degree of association between two variables.  

---

#### **4. Correlation check is used to:**  
⬜ Determine the strength of the linear relationship between two variables  
⬜ Determine the direction of the relationship between two variables  
⬜ Determine the degree of association between two variables  
✅ All of the above  

**Explanation:**  
✅ **Correct Answer: All of the above**—Correlation measures both the strength and direction of the relationship between two variables.  

---

#### **5. A dataset contains a categorical variable with no natural ordering. Which encoding technique would be most appropriate for this variable?**  
⬜ Ordinal Encoding  
✅ Label Encoding  
⬜ Target Guided Ordinal Encoding  
⬜ Covariance  

**Explanation:**  
✅ **Correct Answer: Label Encoding**—Used when categorical variables have no order, assigning a unique integer to each category.  

⬜ **Ordinal Encoding?** Incorrect—used for ordered categories.  
⬜ **Target Guided Ordinal Encoding?** Incorrect—used when categories are strongly related to the target variable.  
⬜ **Covariance?** Incorrect—covariance measures relationships, not encoding.  

---

#### **6. A dataset contains a categorical variable that is strongly associated with the target variable. Which encoding technique would be most appropriate for this variable?**  
⬜ Ordinal Encoding  
⬜ Label Encoding  
✅ Target Guided Ordinal Encoding  
⬜ Covariance  

**Explanation:**  
✅ **Correct Answer: Target Guided Ordinal Encoding**—This technique assigns values to categories based on their relationship with the target variable.  

⬜ **Ordinal Encoding?** Incorrect—used for naturally ordered categories.  
⬜ **Label Encoding?** Incorrect—assigns arbitrary numerical values without considering the target.  
⬜ **Covariance?** Incorrect—measures the relationship between variables but does not encode data.  

---

#### **7. A data scientist is performing feature engineering on a dataset with categorical variables. Which technique can be used to check the degree of association between two categorical variables?**  
⬜ Ordinal Encoding  
⬜ Label Encoding  
✅ Correlation Check  
⬜ Target Guided Ordinal Encoding  

**Explanation:**  
✅ **Correct Answer: Correlation Check**—It helps determine the relationship strength between two variables.  

⬜ **Ordinal Encoding?** Incorrect—used for encoding ordered categories, not checking associations.  
⬜ **Label Encoding?** Incorrect—assigns numbers to categories but does not measure association.  
⬜ **Target Guided Ordinal Encoding?** Incorrect—used for encoding, not measuring relationships.
