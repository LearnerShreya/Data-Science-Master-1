### **Feature Engineering Quiz - 2**  

**6 out of 6 correct**

1. **Which of the following is a feature selection method that selects a subset of features based on their correlation with the target variable?**  

   ✅ **Filter method**  
   ⬜ Wrapper method  
   ⬜ Embedded method  
   ⬜ All of the above  

   **Explanation:**  
   - **Correct Answer:** The filter method ranks features based on statistical metrics like correlation, mutual information, or chi-square tests. It selects features independently of the model, making it computationally efficient.  
   - **Incorrect Options:**  
     - Wrapper method evaluates feature subsets based on model performance, not just correlation.  
     - Embedded method integrates feature selection into the model training process, which is not solely based on correlation.  
     - "All of the above" is incorrect because only the filter method focuses on correlation with the target variable.  

---

2. **Which of the following is a feature selection method that selects a subset of features based on their ability to improve the performance of a given model?**  

   ⬜ Filter method  
   ✅ **Wrapper method**  
   ⬜ Embedded method  
   ⬜ None of the above  

   **Explanation:**  
   - **Correct Answer:** The wrapper method evaluates feature subsets by training and testing the model on each subset. It selects features based on their impact on model performance.  
   - **Incorrect Options:**  
     - Filter method selects features based on statistical metrics, not model performance.  
     - Embedded method combines feature selection with model training but does not explicitly evaluate subsets for performance improvement.  
     - "None of the above" is incorrect because the wrapper method fits the description.  

---

3. **Which of the following is a feature selection method that combines feature selection with the model training process?**  

   ⬜ Filter method  
   ⬜ Wrapper method  
   ✅ **Embedded method**  
   ⬜ None of the above  

   **Explanation:**  
   - **Correct Answer:** The embedded method integrates feature selection into the model training process. Examples include Lasso regression (L1 regularization) and feature importance in decision trees.  
   - **Incorrect Options:**  
     - Filter method is independent of the model and does not combine feature selection with training.  
     - Wrapper method evaluates subsets separately and does not integrate feature selection into training.  
     - "None of the above" is incorrect because the embedded method fits the description.  

---

4. **Which of the following is a disadvantage of the filter method for feature selection?**  

   ⬜ It is model-specific  
   ⬜ It is computationally expensive  
   ✅ **It does not consider the interactions between features**  
   ⬜ It is not based on a statistical metric  

   **Explanation:**  
   - **Correct Answer:** The filter method evaluates features individually, ignoring interactions between them. This can lead to selecting redundant or suboptimal features.  
   - **Incorrect Options:**  
     - Filter method is not model-specific; it is independent of the model.  
     - Filter method is computationally efficient, not expensive.  
     - Filter method is based on statistical metrics, so this option is incorrect.  

---

5. **Which of the following is an advantage of the wrapper method for feature selection?**  

   ⬜ It is computationally efficient  
   ⬜ It is independent of the model used  
   ✅ **It considers the interactions between features**  
   ⬜ It is based on a statistical metric  

   **Explanation:**  
   - **Correct Answer:** The wrapper method evaluates feature subsets by considering their combined impact on model performance, capturing interactions between features.  
   - **Incorrect Options:**  
     - Wrapper method is computationally expensive, not efficient.  
     - Wrapper method is model-dependent, not independent.  
     - Wrapper method is not based on statistical metrics; it relies on model performance.

6. **What is the primary difference between the filter method and the wrapper method for feature selection?**  

   ⬜ The filter method is model-specific, while the wrapper method is independent of the model used.  
   ✅ **The filter method selects features based on their correlation with the target variable, while the wrapper method evaluates subsets of features based on their ability to improve model performance.**  
   ⬜ The filter method considers the interactions between features, while the wrapper method does not.  
   ⬜ The filter method combines feature selection with the model training process, while the wrapper method does not.  

   **Explanation:**  
   - **Correct Answer:** The filter method uses statistical metrics (e.g., correlation) to rank features, while the wrapper method evaluates subsets based on their impact on model performance.  
   - **Incorrect Options:**  
     - Filter method is model-agnostic, while the wrapper method is model-specific.  
     - Filter method does not consider feature interactions, while the wrapper method does.  
     - Embedded method, not the filter method, combines feature selection with model training.
