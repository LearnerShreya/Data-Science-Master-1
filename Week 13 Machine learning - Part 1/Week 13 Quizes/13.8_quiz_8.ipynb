### **1. Which of the following correlation coefficients measures the strength and direction of a linear relationship between two continuous variables?**  

✅ **(A) Pearson correlation coefficient**  
⬜ **(B) Spearman’s rank correlation**  
⬜ **(C) VIF**  
⬜ **(D) None of the above**  

### **Explanation:**  
The **Pearson correlation coefficient** is used to measure the **strength and direction** of a **linear** relationship between two continuous variables. It ranges from **-1 to +1**, where:  
- **-1** indicates a perfect negative correlation,  
- **0** indicates no correlation, and  
- **+1** indicates a perfect positive correlation.  

### **Reason for Incorrect Answers:**  
- **Spearman’s rank correlation** is used for monotonic relationships, not strictly linear ones.  
- **VIF (Variance Inflation Factor)** measures multicollinearity, not correlation.  
- **None of the above** is incorrect as Pearson’s coefficient does measure linear correlation.

### **2. Which correlation coefficient is appropriate to use when the data is ordinal or non-parametric?**  

⬜ **(A) Pearson correlation coefficient**  
✅ **(B) Spearman’s rank correlation**  
⬜ **(C) VIF**  
⬜ **(D) None of the above**  

### **Explanation:**  
**Spearman’s rank correlation** is a **non-parametric** measure of correlation, making it suitable for **ordinal** or **non-parametric** data. It evaluates the **strength and direction** of a **monotonic** relationship between two continuous or ordinal variables. Unlike Pearson’s correlation, it does not assume a linear relationship.  

### **Reason for Incorrect Answers:**  
- **Pearson correlation coefficient** is used for continuous variables in a linear relationship, not ordinal data.  
- **VIF** is not a correlation measure but is used to detect multicollinearity.  
- **None of the above** is incorrect as Spearman’s correlation is the correct choice.

### **3. What is the range of VIF values?**  

⬜ **(A) 0 to 1**  
⬜ **(B) 1 to 10**  
✅ **(C) 0 to infinity**  
⬜ **(D) None of the above**  

### **Explanation:**  
**Variance Inflation Factor (VIF) values range from 0 to infinity.**  
- **A VIF of 1** indicates no multicollinearity.  
- **Higher values (>1)** indicate increasing levels of multicollinearity.  

### **Reason for Incorrect Answers:**  
- **0 to 1** is incorrect because VIF cannot be less than 1 for real data.  
- **1 to 10** is misleading, as VIF can exceed 10.  
- **None of the above** is incorrect because the correct range is **0 to infinity**.

### **4. Which of the following correlation coefficients is more robust to outliers?**  

⬜ **(A) Pearson correlation coefficient**  
✅ **(B) Spearman’s rank correlation**  
⬜ **(C) VIF**  
⬜ **(D) None of the above**  

### **Explanation:**  
**Spearman’s rank correlation** is more **robust to outliers** because it is a **non-parametric measure** that uses **ranks** instead of actual values. Since it focuses on the order rather than magnitude, extreme values do not heavily affect the correlation calculation.  

### **Reason for Incorrect Answers:**  
- **Pearson correlation coefficient** is sensitive to outliers because it directly uses numerical values.  
- **VIF** is not a correlation measure but a multicollinearity detector.  
- **None of the above** is incorrect as Spearman’s correlation is the correct answer.

### **8. A scientist is studying the relationship between the number of hours spent studying and the final exam scores of 50 students. He computes a Pearson correlation coefficient of -0.2. What does this result indicate?**  

⬜ **(A) There is a strong negative correlation between the two variables**  
✅ **(B) There is a weak negative correlation between the two variables**  
⬜ **(C) There is no significant correlation between the two variables**  
⬜ **(D) There is a weak positive correlation between the two variables**  

### **Explanation:**  
The **Pearson correlation coefficient ranges from -1 to 1**, where:  
- **-1** indicates a perfect negative correlation,  
- **0** indicates no correlation, and  
- **1** indicates a perfect positive correlation.  

A coefficient of **-0.2** indicates a **weak negative correlation**, meaning as the number of hours spent studying increases, final exam scores **tend to decrease slightly**, but the relationship is weak.  

### **Reason for Incorrect Answers:**  
- **Strong negative correlation** is incorrect because -0.2 is weak, not strong.  
- **No significant correlation** is incorrect because there is still a weak relationship.  
- **Weak positive correlation** is incorrect because the coefficient is negative.
