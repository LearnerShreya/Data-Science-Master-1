#  Feature Scaling
Feature scaling is a data preprocessing technique used in machine learning to standardize the range of independent variables or features. It ensures that all features contribute equally to the model, preventing certain features with larger ranges from dominating the learning process.

### **Why is Feature Scaling Needed?**
- Some machine learning algorithms (e.g., gradient descent-based models, k-NN, SVM, PCA) perform better when features are on a similar scale.
- It speeds up convergence in gradient-based optimizations.
- It improves model performance by preventing bias toward features with larger magnitudes.

### **Types of Feature Scaling**
1. **Min-Max Scaling (Normalization)**
   - Scales values between a fixed range, usually  \([0,1]\) or \([-1,1]\) $$.
   - Formula:
     $$
     X' = \frac{X - X_{min}}{X_{max} - X_{min}}
     $$
   - Suitable for algorithms that assume bounded data (e.g., neural networks, k-NN).

2. **Standardization (Z-score Normalization)**
   - Centers data around the mean with unit variance.
   - Formula:
     $$
     X' = \frac{X - \mu}{\sigma}
     $$
   - Useful for algorithms that assume normally distributed data (e.g., linear regression, PCA).

3. **Robust Scaling**
   - Uses median and interquartile range to handle outliers.
   - Formula:
     $$
     X' = \frac{X - \text{median}}{\text{IQR}}
     $$
   - Suitable when data contains outliers.

4. **Log Transformation**
   - Reduces skewness in data by applying logarithmic transformation.
   - Formula:
     $$
     X' = \log(X + c)
     $$
   - Helps when data is highly skewed.

### **Which Scaling to Use?**
- **Min-Max Scaling**  When you need values in a fixed range.
- **Standardization**  When data has a normal distribution.
- **Robust Scaling**  When outliers are present.
- **Log Transformation**  When data is skewed.


##  **Unit Vector Scaling (Vector Normalization)**
Unit vector scaling, also known as **vector normalization**, transforms a feature vector to have a **unit norm** (magnitude of 1). This ensures that all features contribute equally in terms of direction but not necessarily in magnitude.

### **Formula**  
For a given feature vector \( X \), the normalized vector \( X' \) is computed as:

$$
X' = \frac{X}{||X||}
$$

where \( ||X|| \) is the **Euclidean norm (L2 norm)**, defined as:

$$
||X|| = \sqrt{x_1^2 + x_2^2 + ... + x_n^2}
$$

Alternatively, other norms like **L1 norm** (sum of absolute values) or **L norm** (max absolute value) can be used.

### **When to Use Unit Vector Scaling?**
- When the direction of the feature vector matters more than its magnitude (e.g., text classification, cosine similarity in NLP).
- When features need to be comparable on the same scale (e.g., k-NN, SVM with cosine similarity).
- When working with sparse data where standard scaling might distort results.

### **Python Implementation**
Using **scikit-learn**:
```python
from sklearn.preprocessing import normalize
import numpy as np

X = np.array([[4, 1, 2], [1, 3, 9], [7, 5, 6]])
X_normalized = normalize(X, norm='l2')  # 'l2' for unit vector scaling
print(X_normalized)
```


##  **Feature Scaling: Normalization**  
Normalization is a feature scaling technique that **rescales** data to a fixed range, usually **[0,1]** or **[-1,1]**. It ensures that all features contribute equally to the model and improves convergence in optimization algorithms.  

### **Formula for Min-Max Normalization**  
For a given feature \( X \), normalization is done using:  

$$  
X' = \frac{X - X_{min}}{X_{max} - X_{min}}  
$$  

where:  
$$  
X = \text{Original feature value}  
$$  

$$  
X_{min} = \text{Minimum value in the feature}  
$$  

$$  
X_{max} = \text{Maximum value in the feature}  
$$  

$$  
X' = \text{Normalized value in the range } [0,1]  
$$  

---

### **When to Use Normalization?**  
 When features have **different scales** and need to be **bounded** (e.g., neural networks, k-NN, SVM).  
 When feature values are **not normally distributed** (e.g., skewed data).  
 When working with distance-based models like **k-means clustering** or **PCA**.  

---

### **Python Code Example (Min-Max Scaling)**  
Using **scikit-learn**:
```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Sample Data
X = np.array([[10], [20], [30], [40], [50]])

# Apply Min-Max Normalization
scaler = MinMaxScaler(feature_range=(0, 1))
X_normalized = scaler.fit_transform(X)

print("Original Data:\n", X)
print("Normalized Data:\n", X_normalized)
```

---

### **Normalization vs Standardization**  

| Feature Scaling | Formula | Range | Use Case |
|---|---|---|---|
| **Normalization (Min-Max Scaling)** | $$ X' = \frac{X - X_{min}}{X_{max} - X_{min}} $$ | [0,1] or [-1,1] | When data is **not normally distributed** |
| **Standardization (Z-score Normalization)** | $$ X' = \frac{X - \mu}{\sigma} $$ | Mean = 0, Std = 1 | When data follows a **normal distribution** |
